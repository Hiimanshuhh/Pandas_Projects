```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np 
```

```python
x = np.linspace(0,1,11)
```

```python
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'r')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Graph")
plt.tight_layout()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.subplot(1,2,2)
plt.plot(x**2,y,'g')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Graph")
plt.tight_layout()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes = fig.add_axes([0.30,0.30,0.4,0.4])
axes.plot(y,x)
```

```python
fig , axes = plt.subplots(1,4)
for ax in axes:
    ax.plot(x,y)
plt.tight_layout()
```

```python
fig = plt.figure(figsize =(6,6),dpi=100)
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,label='x-line')
axes.plot(x,y**2,label='y-line')
axes.legend(loc=0)
```

```python
fig = plt.figure()
axe = fig.add_axes([0.1,0.1,0.8,0.8])
axe.plot(x,y,color="purple",lw=3,ls="--",alpha=0.5,marker='o',markersize=20,markerfacecolor="pink",markeredgewidth=9,markeredgecolor='Green')
```
